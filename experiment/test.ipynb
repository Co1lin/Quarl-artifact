```python
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, transpile, qasm
from qiskit.test.mock import FakeBoeblingen
backend = FakeBoeblingen()
```

```python
file = 'barenco_tof_3'
before = QuantumCircuit.from_qasm_file('../circuit/nam-circuits/qasm_files/' + file + '_before.qasm')
before.draw(output='mpl')
```

```python
after = QuantumCircuit.from_qasm_file('../circuit/nam-circuits/qasm_files/' + file + '_after_light.qasm')
after.draw(output='mpl')
```

```python
optimized_circ = transpile(after, basis_gates=['u1', 'u2', 'u3', 'cx'], optimization_level=3)
print(optimized_circ.count_ops())
print(optimized_circ.depth())
optimized_circ.draw(output='mpl')
```
